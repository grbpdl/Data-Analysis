We can have data in our local machine or online.There are various formats for a data set: .csv, .json, .xlsx etc.We use pandas library to read the datas.

```python
import pandas as pd
import numpy as np
```

 we  can use read_csv() to load the data into a dataframe.

```python
df=pd.read_csv('auto.csv')
```

to look for first five rows from dataframe we use head()

```python
df.head()
```

```python
df.tail(10) #shows bottom 10 rows
```

# Setting the headers to dataframe

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
df.columns
```

```python
df.head(5)
```

we should replace the question mark with some more meaningful thing like NaN

```python
df1=df.replace('?',np.NaN)
```

```python
df1.head(4)
```

we can also drop the rows having missing values from certain column like price

```python
df=df1.dropna(subset=["price"], axis=0) #Here, axis=0 means that the contents along the entire row will be dropped wherever the entity 'price' is found to be NaN
df.head(5)
```

# saving pandas dataset into csv

```python
df.to_csv("automobile.csv", index=False) #index = False means the row names will not be written.
```
